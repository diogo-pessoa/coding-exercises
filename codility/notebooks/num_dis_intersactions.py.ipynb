```python
import matplotlib.pyplot as plt
import numpy as np

# Example radii and centers based on your array A example
A = [1, 5, 2, 1, 4, 0]
N = len(A)

# Create the figure and axis
fig, ax = plt.subplots()

# Iterate over each disc to plot
for i, radius in enumerate(A):
    circle = plt.Circle((i, 0), radius, color=plt.cm.viridis(i / N), alpha=0.5, label=f'Disc {i}', fill=False)
    ax.add_artist(circle)

# Setting the limits of the plot
# Find the maximum radius to set the limits
max_radius = max(A)
ax.set_xlim(-max_radius, N + max_radius)
ax.set_ylim(-max_radius, max_radius)

# Adding equal aspect to make the circles look like circles
ax.set_aspect('equal')

# Adding the legend outside the plot to the right
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Plot of Discs with Legends')
plt.show()
```
