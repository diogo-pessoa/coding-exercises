```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from functools import partial

# Initial data
max_value = 40
changing_values = [0]

# Figure setup
fig, ax = plt.subplots()
line, = ax.plot([], [], 'r-')
ax.set_xlim(0, max_value)
ax.set_ylim(0, max_value)

# Animation function (using partial for line update)
def update_line(frame, line, changing_values):
    current_value = max_value - frame
    changing_values[0] = frame
    line.set_data([current_value, current_value], [0, changing_values[0]])
    return line,

# Create partial function
update_line_partial = partial(update_line, line=line, changing_values=changing_values)

# Animation (using the partial function)
ani = FuncAnimation(fig, update_line_partial, frames=range(max_value + 1), interval=1000, blit=True, repeat=False)

plt.show()
```
